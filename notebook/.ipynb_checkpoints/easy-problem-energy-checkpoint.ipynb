```python
import numpy as np
import sys
sys.path.append("../code")
from basic import *
from algo_fix_season import *
from algo import *
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2
```

```python
alpha1 = 0.1
alpha2 = 0.1
alpha3 = 0.1
lambda_ = 10000
init = 'pre'
normalization = 'none'
method = 'active'
```

```python
latent_dimension = 3
tensor, homeids = get_tensor(2015, 'filter')
```

```python
AS_none = {}
AS_all = {}
for fold_num in range(5):
    AS_none[fold_num] = {}
    AS_all[fold_num] = {}
    train, test, tr_homes, tt_homes = get_train_test(tensor, homeids, fold_num=fold_num)
    for random_seed in range(5):
        inst_tensor, init_list = get_instrumented_data(2015, 'filter', tr_homes, 10, random_seed=random_seed)
        h, app, season = factorization(inst_tensor, num_latent=latent_dimension, dis=True, random_seed=random_seed)
        AS_none[fold_num][random_seed] = ActiveSensing(train_tensor=train, test_tensor=test, init_list=init_list, pre_app=app, pre_season=season, 
                                method=method, latent_dimension=latent_dimension, alpha1=alpha1, alpha2=alpha2, alpha3=alpha3, 
                                lambda_=lambda_, init=init, uncertainty='equal', k=0, 
                                normalization=normalization, random_seed=random_seed)
        AS_all[fold_num][random_seed] = ActiveSensing(train_tensor=train, test_tensor=test, init_list=init_list, pre_app=app, pre_season=season, 
                                method='all', latent_dimension=latent_dimension, alpha1=alpha1, alpha2=alpha2, alpha3=alpha3, 
                                lambda_=lambda_, init=init, uncertainty='equal', k=0, 
                                normalization=normalization, random_seed=random_seed)
        for t in range(12):
            AS_none[fold_num][random_seed].select(t)
            AS_none[fold_num][random_seed].update_ALS_2(t, 5)
            AS_all[fold_num][random_seed].select(t)
            AS_all[fold_num][random_seed].update_ALS_2(t, 5)
```

```python
for random_seed in range(20):
    df1 = pd.DataFrame(AS_all[random_seed].test_errors).mean()
    df2 = pd.DataFrame(AS_none[random_seed].test_errors).mean()
    df = pd.concat([df1, df2], axis=1)
    df.plot(kind='bar')
```
