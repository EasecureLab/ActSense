```python
import numpy as np
import sys
sys.path.append("../code/")
from structure import *
from algo import *
from basic import *
from analysis import *
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from autograd.numpy import linalg as LA
import pickle
import itertools
from itertools import product
import calendar
from sklearn.metrics import mean_squared_error
%load_ext autoreload
%autoreload 2
import warnings
warnings.filterwarnings("ignore")
```

### Set parameters

```python
year = 2015
dataset = 'artificial'
init = 'pre'
k = 5
latent_dimension = 3
if dataset == 'artificial':
    order = APPLIANCE_ORDER_OBSERVED
else:
    order = APPLIANCE_ORDER_MISSING
```

### get the results

```python
result = {}

# without regularization
# for param in ['random', 'qbc']:
#     result[param] = get_active_results(method=param, reg=False)
# for param in ['one', 'equal', 'weighted', 'next']:
#     result[param] = get_active_results(uncertainty=param, season='updated', reg=False)
# for param in ['updated']:
#     result[param] = get_active_results(season=param, reg=False)
# with regularization
for param in ['random', 'qbc']:
    result[param + " reg"] = get_active_results(method=param, season='update', reg=True)
for param in ['one', 'equal', 'weighted', 'next']:
    result[param + " reg"] = get_active_results(uncertainty=param, season='updated', reg=True)
for param in ['updated']:
    result[param + " reg"] = get_active_results(season=param, reg=True)
```

### Prepare the structure

```python
error = {}
for e_type in ['overall', 'app']:
    error[e_type] = {}
    for t_type in ['accu', 'month']:
        error[e_type][t_type] = {}
        for param in result.keys():
            error[e_type][t_type][param] = {}
            for idx, appliance in enumerate(order[1:]):
                error[e_type][t_type][param][appliance] = {}
                for t in range(12):
                    error[e_type][t_type][param][appliance][t] = {}
                    for random_seed in range(10):
                        error[e_type][t_type][param][appliance][t][random_seed] = {}
```

### get the prediction error

```python
tensor, homeidx = get_tensor(year, dataset)
mask_tensor = ~np.isnan(tensor)

for param in result.keys():
    for t in range(12):
        # random method
        for idx, appliance in enumerate(order[1:]):
            idx += 1
            for random_seed in range(10):
                # for active method
                predApp = result[param]['prediction'][random_seed][t][:, idx, :] # tensor
                
                # accumulated appliance error
                appAccuPred = predApp[:, :(t+1)]
                error['app']['accu'][param][appliance][t][random_seed] = np.sqrt(mean_squared_error(appAccuPred[mask_tensor[:, idx, :(t+1)]], 
                                                                                               tensor[:, idx, :(t+1)][mask_tensor[:, idx, :(t+1)]]))
                # appliance error in current month
                appMonthPred = predApp[:, t]
                error['app']['month'][param][appliance][t][random_seed] = np.sqrt(mean_squared_error(appMonthPred[mask_tensor[:, idx, t]], 
                                                                                                tensor[:, idx, t][mask_tensor[:, idx, t]]))
```

### plot the accumulated error for each appliance

```python
params = result.keys()
params = ['one reg', 'equal reg', 'weighted reg', 'random reg', 'qbc reg']

mean_error = {}
# for param in result.keys():
for param in params:
    mean_error[param] = {}
    for appliance in order[1:]:
        mean_error[param][appliance] = pd.DataFrame(error['app']['accu'][param][appliance]).mean(axis=0)
        
fig, axes = plt.subplots(ncols=int((len(order)-1)/2), nrows=2, figsize=(15,5))
ax = axes.flatten()
for idx, appliance in enumerate(order[1:]):
    df = {}
    for param in params:
        df[param] = mean_error[param][appliance]
    pd.DataFrame(df).plot(ax=ax[idx], kind='bar', title=appliance)
    ax[idx].legend_.set_visible(False)
ax[idx].legend_.set_visible(True)
ax[idx].legend(loc='lower right')
plt.subplots_adjust(hspace=0.5)

AccumulatedError = {}
for param in params:
    AccumulatedError[param] = pd.DataFrame(mean_error[param]).mean(axis=1)

df = pd.DataFrame(AccumulatedError).rename(columns={'qbc':'Query by Committee', 'updated':'UCB_uncertainty', 'random':'Random', 'active_error':'UCB_uncertainty + error'})
df = df.rename(index={0:'Jan', 1:'Feb', 2:'Mar', 3:'Apr', 4:'May', 5:'Jun', 6:'Jul', 7:'Aug', 8:'Sep', 9:'Oct', 10:'Nov', 11:'Dec'})   
# ax = pd.DataFrame(AccumulatedError).plot(kind='bar', title='Overall Accumulated Error')

ax = df.plot(kind='bar', title='Overall Accumulated Error')
ax.legend(loc='lower right', bbox_to_anchor=[1.5, 0.5])
ax.set_ylabel("RMSE")
ax.set_xlabel("Month")
```

### plot monthly error for each appliance

```python
params = result.keys()
params = ['one reg', 'equal reg', 'weighted reg', 'random reg', 'qbc reg']

mean_error = {}
for param in params:
    mean_error[param] = {}
    for appliance in order[1:]:
        mean_error[param][appliance] = pd.DataFrame(error['app']['month'][param][appliance]).mean(axis=0)
        
fig, axes = plt.subplots(ncols=int((len(order)-1)/2), nrows=2, figsize=(15,5))
ax = axes.flatten()
for idx, appliance in enumerate(order[1:]):
    df = {}
    for param in params:
        df[param] = mean_error[param][appliance]
    pd.DataFrame(df).plot(ax=ax[idx], kind='bar', title=appliance)
    ax[idx].legend_.set_visible(False)
ax[idx].legend_.set_visible(True)
plt.subplots_adjust(hspace=0.5)

AccumulatedError = {}
for param in params:
    AccumulatedError[param] = pd.DataFrame(mean_error[param]).mean(axis=1)
ax = pd.DataFrame(AccumulatedError).plot(kind='bar', title='Monthly Error')
ax.legend(loc='lower right', bbox_to_anchor=(1, 1))

# get the improvement from the random method
```
