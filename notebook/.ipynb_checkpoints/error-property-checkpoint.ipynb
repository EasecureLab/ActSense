```python
train_agg_mean = train[:, 0, :].mean(axis=1)
sort_index = np.argsort(train_agg_mean)
```

```python
def reorder(arr, index): 
    temp = np.zeros(arr.shape)
    for i in range(len(index)): 
        temp[i] = arr[index[i]] 
    return temp
```

```python
fig, axes = plt.subplots(nrows=11, ncols=4, figsize=(15, 30))
ax = axes.flatten()

for t in range(11):
    active_one.select(t)
    active_one.update_als_update_season(t, 5)
    # get h_circle_s and a_circle_s for current month
    hs_1 = np.zeros((active_one.num_train, latent_dimension))
    hs_2 = np.zeros((active_one.num_train, latent_dimension))
    as_1 = np.zeros((active_one.num_appliance, latent_dimension))
    as_2 = np.zeros((active_one.num_appliance, latent_dimension))
    for i in range(active_one.num_train):
        hs_1[i] = np.multiply(active_one.homes[i].h, active_one.seasons[t].s).T
        tmp = 0
        for k in range(t+1, 11):
#             print(active_one.homes[i].h)
#             print(active_one.seasons[k].s)
            tmp += np.multiply(active_one.homes[i].h, active_one.seasons[k].s).T
        hs_2[i] = tmp
#     for j in range(active_one.num_appliance):
#         as_1[j] = np.multiply(active_one.apps[j].a, active_one.seasons[t].s).T
#         as_2[j] = np.multiply(active_one.apps[j].a, active_one.seasons[t + 1].s).T
        
    # reorder the values with the sorted index.
    new_train_home_matrix = reorder(active_one.train_home_matrix[t], sort_index)
    new_hs_1 = reorder(hs_1, sort_index)
    new_hs_2 = reorder(hs_2, sort_index)
    
    # plot the results
    pd.DataFrame(new_train_home_matrix).plot(ax=ax[t*4], title='home factor')
    pd.DataFrame(new_hs_1).plot(ax=ax[1+t*4], title='home * S_current')
    pd.DataFrame(new_hs_2).plot(ax=ax[2+t*4], title='home * sum(S_future)')
    pd.DataFrame(active_one.season_matrix[t]).plot(ax=ax[3+t*4], title='season factor')
    ax[t*4].get_legend().remove()
    ax[t*4+1].get_legend().remove()
    ax[t*4+2].get_legend().remove()

    ax[3+t*4].xaxis.set_ticks(np.arange(0, 12, 1))
    ax[t*4].set_ylabel("Month {}".format(t))
    
plt.subplots_adjust(hspace=0.5)
```

```python
def calculate_selected_appliance(sp, k):
    selected_app = np.zeros((12, 7))
#     for t in range(1, 12):
#         selected_app[t] = {}
#         for app in order:
#             selected_app[t][app] = 0
    for t in range(1, 12):
        for i in range(k):
#                     print(random_seed, fold_num, t, i)
            app_idx = sp[t][i][1]
#             app = order[app_idx]
            selected_app[t][app_idx] += 1
    return selected_app
```
