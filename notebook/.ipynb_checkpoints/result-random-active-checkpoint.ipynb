### get the results of active selection

```python
k = 5
def get_active_results(year=2015, dataset='artificial', method='active', alpha1=0.1, alpha2=0.1, alpha3=0.0, normalization='none', uncertainty='one', init='pre', k=5):
    directory = "../data/fixed_season/{}/{}/{}/".format(method, year, dataset)
    # load prediction
    file_name = "pred-{}-{}-{}-{}-{}-{}-{}-{}".format(init, normalization, uncertainty, alpha1, alpha2, alpha3, k, latent_dimension)
    file = open(directory + file_name, 'rb')
    pred = pickle.load(file)
    file.close()
    
    file_name = "sp-{}-{}-{}-{}-{}-{}-{}-{}".format(init, normalization, uncertainty, alpha1, alpha2, alpha3, k, latent_dimension)
    file = open(directory + file_name, 'rb')
    sp = pickle.load(file)
    file.close()
    # load the appliance matrix
    file_name = "app-matrix-{}-{}-{}-{}-{}-{}-{}-{}".format(init, normalization, uncertainty, alpha1, alpha2, alpha3, k, latent_dimension)
    file = open(directory + file_name, 'rb')
    app_matrix = pickle.load(file)
    file.close()
    # load the home matrix
    # random_seed, fold_num, season_id
    file_name = "test-matrix-{}-{}-{}-{}-{}-{}-{}-{}".format(init, normalization, uncertainty, alpha1, alpha2, alpha3, k, latent_dimension)
    file = open(directory + file_name, 'rb')
    test_home = pickle.load(file)
    file.close()
    file_name = "train-matrix-{}-{}-{}-{}-{}-{}-{}-{}".format(init, normalization, uncertainty, alpha1, alpha2, alpha3, k, latent_dimension)
    file = open(directory + file_name, 'rb')
    train_home = pickle.load(file)
    file.close()
    # load the season matrix
    # random_seed, fold_num, season_id
    file_name = "season-matrix-{}-{}-{}-{}-{}-{}-{}-{}".format(init, normalization, uncertainty, alpha1, alpha2, alpha3, k, latent_dimension)
    file = open(directory + file_name, 'rb')
    season = pickle.load(file)
    file.close()
    
    return pred, sp, app_matrix, test_home, train_home, season
```

```python
a_pred, a_sp, a_a, a_test, a_train, a_season = get_active_results(method='active')
c_pred, c_sp, c_a, c_test, c_train, c_season = get_active_results(method='cluster_3')
a_pred_2, a_sp_2, a_a_2, a_test_2, a_train_2, a_season_2 = get_active_results(method='active_error')
```

### using RMSE as the evaluation metric

```python
mask_tensor = ~np.isnan(tensor)
```

```python
def get_app_error(predApp, gt, t):
    mask_tensor = ~np.isnan(gt)
    appAccuPred = predApp[:, :(t+1)]
    app_error = np.sqrt(mean_squared_error(appAccuPred[mask_tensor[:, idx, :(t+1)]], gt[:, idx, :(t+1)][mask_tensor[:, idx, :(t+1)]]))
    appMonthPred = predApp[:, t]
    month_error = np.sqrt(mean_squared_error(appMonthPred[mask_tensor[:, idx, t]], gt[:, idx, t][mask_tensor[:, idx, t]]))
    return app_error, month_error
```

```python
for t in range(12):
    # random method
    for idx, appliance in enumerate(order[1:]):
        idx += 1
        for random_seed in range(10):
            # for random method
            predApp = r_pred[random_seed][t][:, idx, :] # tensor
            appAccuError['random'][appliance][t][random_seed], appMonthError['random'][appliance][t][random_seed] = get_app_error(predApp, tensor, t)
            
            predApp = a_pred[random_seed][t][:, idx, :] # tensor
            appAccuError['active'][appliance][t][random_seed], appMonthError['active'][appliance][t][random_seed] = get_app_error(predApp, tensor, t)

            predApp = a_pred_2[random_seed][t][:, idx, :] # tensor
            appAccuError['active_2'][appliance][t][random_seed], appMonthError['active_2'][appliance][t][random_seed] = get_app_error(predApp, tensor, t)

            predApp = c_pred[random_seed][t][:, idx, :] # tensor
            appAccuError['cluster_3'][appliance][t][random_seed], appMonthError['cluster_3'][appliance][t][random_seed] = get_app_error(predApp, tensor, t)

#             predApp = a_pred_4[random_seed][t][:, idx, :] # tensor
#             appAccuError['active_4'][appliance][t][random_seed], appMonthError['active_4'][appliance][t][random_seed] = get_app_error(predApp, tensor, t)

#             predApp = a_pred_5[random_seed][t][:, idx, :] # tensor
#             appAccuError['active_5'][appliance][t][random_seed], appMonthError['active_5'][appliance][t][random_seed] = get_app_error(predApp, tensor, t)
            
            predApp = q_pred[random_seed][t][:, idx, :] # tensor
            appAccuError['qbc'][appliance][t][random_seed], appMonthError['qbc'][appliance][t][random_seed] = get_app_error(predApp, tensor, t)
```

```python
mask_tensor = ~np.isnan(random_test.train_tensor)
cError = {}
for i in range(4):
    cError[i] = {}
    for app_idx in range(1, 7):
        cError[i][app_idx] = {}
        
#     home_factor = test.train_home_matrix[11][kmeans.labels_ == i]
        pred = np.einsum('hr, ar, sr -> has', active_test.train_home_matrix[11], active_test.app_matrix[11], active_test.season_matrix[11])
        rmse = np.sqrt(mean_squared_error(pred[kmeans.labels_ == i, app_idx][mask_tensor[kmeans.labels_ == i, app_idx]],
                                         active_test.train_tensor[kmeans.labels_ == i, app_idx][mask_tensor[kmeans.labels_ == i, app_idx]]))
#         print(i, rmse)
        cError[i][app_idx]['active'] = rmse
        pred = np.einsum('hr, ar, sr -> has', random_test.train_home_matrix[11], random_test.app_matrix[11], random_test.season_matrix[11])
        rmse = np.sqrt(mean_squared_error(pred[kmeans.labels_ == i, app_idx][mask_tensor[kmeans.labels_ == i, app_idx]],
                                         random_test.train_tensor[kmeans.labels_ == i, app_idx][mask_tensor[kmeans.labels_ == i, app_idx]]))
#         print(i, rmse)
        cError[i][app_idx]['random'] = rmse
    
        pred = np.einsum('hr, ar, sr -> has', active_test_2.train_home_matrix[11], active_test_2.app_matrix[11], active_test_2.season_matrix[11])
        rmse = np.sqrt(mean_squared_error(pred[kmeans.labels_ == i, app_idx][mask_tensor[kmeans.labels_ == i, app_idx]],
                                         active_test_2.train_tensor[kmeans.labels_ == i, app_idx][mask_tensor[kmeans.labels_ == i, app_idx]]))
#         print(i, rmse)
        cError[i][app_idx]['active_2'] = rmse
```
