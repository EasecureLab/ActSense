```python
import numpy as np
import sys
sys.path.append("../code/")
from structure import *
from algo import *
from algo_fix_season import *
from basic import *
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from autograd.numpy import linalg as LA
import pickle

year = 2015
fold_num = 0
random_seed = 9
method = 'active'
latent_dimension = 3
normalization = 'none'
uncertainty = 'one'
alpha1 = 0.1
alpha2 = 0.1
alpha3 = 0.1
lambda_ = 10000
init = 'pre'
k = 5

tensor, homeids = get_tensor(year, 'artificial')
train, test, tr_homes, tt_homes = get_train_test(tensor, homeids, fold_num=fold_num)

# without any instrumented houses, and season factors are learned by the aggregate readings
init_list = None
pre_tensor, homeids = get_tensor(year-1, 'artificial')
pre_tensor[:, 1:] = np.NaN
T = np.ones(12).reshape(-1, 1)
h, app, season = factorization(pre_tensor, num_latent=latent_dimension, dis=True, random_seed=random_seed, T_known=T)

active_test = ActiveSensing_fix(train_tensor=train, test_tensor=test, init_list=None, 
                         pre_app=None, pre_season=season, method='cluster_3', latent_dimension=latent_dimension, 
                         alpha1=alpha1, alpha2=alpha2, alpha3=alpha3, lambda_=lambda_, init=init, 
                         uncertainty=uncertainty, k=k, normalization=normalization, random_seed=random_seed)
random_test = ActiveSensing_fix(train_tensor=train, test_tensor=test, init_list=None, 
                         pre_app=None, pre_season=season, method='cluster_4', latent_dimension=latent_dimension, 
                         alpha1=alpha1, alpha2=alpha2, alpha3=alpha3, lambda_=lambda_, init=init, 
                         uncertainty=uncertainty, k=k, normalization=normalization, random_seed=random_seed)
```

```python
for t in range(12):
    active_test.select(t)
    active_test.update_als_random(t, 5)
    random_test.select(t)
    random_test.update_als_random(t, 5)
```

```python
mask_tensor = ~np.isnan(random_test.train_tensor)
cError = {}
for i in range(4):
    cError[i] = {}
    for app_idx in range(1, 7):
        print(i, app_idx)
        cError[i][app_idx] = {}
        
#     home_factor = test.train_home_matrix[11][kmeans.labels_ == i]
        pred = np.einsum('hr, ar, sr -> has', active_test.train_home_matrix[11], active_test.app_matrix[11], active_test.season_matrix[11])
        rmse = np.sqrt(mean_squared_error(pred[kmeans.labels_ == i, app_idx][mask_tensor[kmeans.labels_ == i, app_idx]],
                                         active_test.train_tensor[kmeans.labels_ == i, app_idx][mask_tensor[kmeans.labels_ == i, app_idx]]))
#         print(i, rmse)
        cError[i][app_idx]['active'] = rmse
        pred = np.einsum('hr, ar, sr -> has', random_test.train_home_matrix[11], random_test.app_matrix[11], random_test.season_matrix[11])
        rmse = np.sqrt(mean_squared_error(pred[kmeans.labels_ == i, app_idx][mask_tensor[kmeans.labels_ == i, app_idx]],
                                         random_test.train_tensor[kmeans.labels_ == i, app_idx][mask_tensor[kmeans.labels_ == i, app_idx]]))
#         print(i, rmse)
        cError[i][app_idx]['random'] = rmse
```

```python
clusterErr = {}
for cluster in range(5):
    clusterErr[cluster] = {}
    for appliance in APPLIANCE_ORDER_MISSING[1:]:
        clusterErr[cluster][appliance] = {}
        clusterErr[cluster][appliance]['active'] = {}
        clusterErr[cluster][appliance]['random'] = {}
# cluster = 4
for cluster in range(5):
    for t in range(12):
        clusterActivePred = np.einsum('hr, ar, sr -> has', 
                                      active_test.train_home_matrix[t], 
                                      active_test.app_matrix[t], 
                                      active_test.season_matrix[t])
        clusterRandomPred = np.einsum('hr, ar, sr -> has', 
                                      random_test.train_home_matrix[t], 
                                      random_test.app_matrix[t], 
                                      random_test.season_matrix[t])
        for app_idx, appliance in enumerate(APPLIANCE_ORDER_MISSING[1:]):
            app_idx += 1
            rmse = np.sqrt(mean_squared_error(clusterActivePred[kmeans.labels_ == cluster, app_idx][mask_tensor[kmeans.labels_ == cluster, app_idx]],
                                             active_test.train_tensor[kmeans.labels_ == cluster, app_idx][mask_tensor[kmeans.labels_ == cluster, app_idx]]))
            clusterErr[cluster][appliance]['active'][t] = rmse
            rmse = np.sqrt(mean_squared_error(clusterRandomPred[kmeans.labels_ == cluster, app_idx][mask_tensor[kmeans.labels_ == cluster, app_idx]],
                                             active_test.train_tensor[kmeans.labels_ == cluster, app_idx][mask_tensor[kmeans.labels_ == cluster, app_idx]]))
            clusterErr[cluster][appliance]['random'][t] = rmse
        
```
