```python
import numpy as np
import sys
sys.path.append("../code/")
from structure import *
from algo import *
from basic import *
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from autograd.numpy import linalg as LA
import pickle
import itertools
from itertools import product
import calendar
```

```python
year = 2015
params = {'method': ['active', 'random', 'oracle'],
          'init': ['pre'],
          'uncertainty': ['one'],
          'normalization': ['none'],
          'alpha1': [0.1],
          'alpha2': [0.1],
          'alpha3': [0.1],
          'k': [5],
          'latent_dimension':[3]}
```

```python
directory = "../data/fixed_season/{}/".format(year)
```

```python
result = {}
for method in ['active', 'random', 'oracle']:
    result[method] = {}
    for uncertainty in ['one']:
        result[method][uncertainty] = {}
        for alpha1 in [0.1]:
            result[method][uncertainty][alpha1] = {}
            for alpha2 in [0.1]:
                result[method][uncertainty][alpha1][alpha2] = {}
                for alpha3 in [0.1]:
                    result[method][uncertainty][alpha1][alpha2][alpha3] = {}
                    for k in [5]:
                        result[method][uncertainty][alpha1][alpha2][alpha3][k] = {}
                        for random_seed in range(10):
                            result[method][uncertainty][alpha1][alpha2][alpha3][k][random_seed] = {}
```

```python
for vals in product(*params.values()):
    param = dict(zip(params, vals))
    for random_seed in range(10):
        method = param['method']
        init = param['init']
        normalization = param['normalization']
        uncertainty = param['uncertainty']
        alpha1 = float(param['alpha1'])
        alpha2 = float(param['alpha2'])
        alpha3 = float(param['alpha3'])
        k = param['k']
        error_file = "error-{}-{}-{}-{}-{}-{}-{}-{}-3-{}".format(method, init, normalization, uncertainty, alpha1, alpha2, alpha3, k, random_seed)
        file_name = directory + error_file
        error_file = open(file_name, 'rb')
        result[method][uncertainty][alpha1][alpha2][alpha3][k][random_seed] = pickle.load(error_file)
```
