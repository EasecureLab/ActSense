```python
import numpy as np
import sys
sys.path.append("../code/")
from structure import *
from algo import *
from basic import *
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from autograd.numpy import linalg as LA
import pickle
import os
from itertools import product
```

```python
params = {'method': ['active', 'random'],
          'init': ['random'],
          'uncertainty': ['one'],
          'normalization': ['none', 'zscore'],
          'lambda1': [1000.0], 
          'lambda2': [1000.0], 
          'lambda3': [1.0],
          'init_ob': [10], 
          'latent_dimension': [3],
          'k': [5]}
```

```python
c = 0
for vals in product(*params.values()):
    param = dict(zip(params, vals))
    if param['method'] == 'random' and param['normalization'] != 'none':
        print(">>>SKIP: No normalization for random selection")
        continue
    if param['init'] == 'random' and param['uncertainty'] != 'one':
        print('>>>SKIP: No future uncertainty for random init')
        continue
    error = 0
    c += 1
    if c%200 == 0:
        print(c)
    for fold_num in range(10):
        for random_seed in range(5):
            filename = "../data/{}/{}/{}/{}/obj-{}-{}-0.1-0.1-0.1-{}-{}-{}-{}-{}-{}.obj".format(param['method'], 
                                                                                                    param['init'], 
                                                                                                    param['uncertainty'], 
                                                                                                    param['normalization'], 
                                                                                                    fold_num, 
                                                                                                    param['latent_dimension'],
                                                                                                    param['lambda1'], 
                                                                                                    param['lambda2'], 
                                                                                                    param['lambda3'],
                                                                                                    param['init_ob'], 
                                                                                                    param['k'], 
                                                                                                    random_seed)
            print(filename)
            try:
                file = open(filename, 'rb')
                m = pickle.load(file, encoding='latin1')
                file.close()
#                 if fold_num < 6:
#                     error += pd.DataFrame(m.test_errors).values
#                 else:
                error += pd.DataFrame(m).values
                print(fold_num, random_seed, pd.DataFrame(m).values.mean())
            except:
#                 print(filename)
                print("python run.py {} {} {} 0.1 0.1 0.1 {} {} {} {} {} {} {} {} {} &".format(fold_num,
                                                                                             param['method'],
                                                                                             param['latent_dimension'],
                                                                                             param['lambda1'],
                                                                                             param['lambda2'],
                                                                                             param['lambda3'],
                                                                                             param['init'],
                                                                                             param['init_ob'],
                                                                                             param['uncertainty'],
                                                                                             param['k'],
                                                                                             param['normalization'],
                                                                                             random_seed))
                continue
    error = error/50
    save_dir = "../data/test_error/{}/{}/{}/{}/{}/{}/".format(param['method'], param['init'], param['uncertainty'], 
                                                              param['normalization'], param['init_ob'], param['k'])
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    save_name = "error-{}-0.1-0.1-0.1-{}-{}-{}".format(param['latent_dimension'], param['lambda1'], 
                                                       param['lambda2'], param['lambda3'])

    name = save_dir+save_name
    print(c, name)
#     print(error)
    file = open(name, 'wb')
    pickle.dump(error, file)
    file.close()
```
