```python
for t in range(12):
    test.select(t)
    test.update_als_random(t, 5)
```

```python
mask_tensor = ~np.isnan(test.train_tensor)
for i in range(5):
#     home_factor = test.train_home_matrix[11][kmeans.labels_ == i]
    pred = np.einsum('hr, ar, sr -> has', test.train_home_matrix[11], test.app_matrix[11], test.season_matrix[11])
    rmse = np.sqrt(mean_squared_error(pred[kmeans.labels_ == i][mask_tensor[kmeans.labels_ == i]],
                                     test.train_tensor[kmeans.labels_ == i][mask_tensor[kmeans.labels_ == i]]))
    print(i, rmse)
```
