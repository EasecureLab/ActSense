```python
import numpy as np
import sys
sys.path.append("../code/")
from structure import *
from algo import *
from basic import *
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from autograd.numpy import linalg as LA
import pickle
```

```python
tensor = get_tensor(2015, 'missing')
# train, test = get_train_test(tensor, fold_num=)
```

```python
app = None
season = None
latent_dimension = 3
alpha1 = 0.1
alpha2 = 0.1
alpha3 = 0.1
lambda1 = 1000
lambda2 = 1000
lambda3 = 1
method = 'active'
init = 'random'
init_ob = 10
uncertainty = 'one'
normalization = 'none'
k = 5
random_seed = 0
```

```python
def active_sensing(method='active', latent_dimension=3, alpha1=0.1, alpha2=0.1, alpha3=0.1, 
                           lambda1=1, lambda2=1, lambda3=1, init='random', init_ob=0, uncertainty='one',
                           k=5, normalization='none', random_seed=0):
    AS = {}
    for fold_num in range(10):
        print("Fold: ", fold_num)
        train, test = get_train_test(tensor, fold_num=fold_num)
        AS[fold_num] = ActiveSensing(train_tensor=train, 
                test_tensor=test, 
                pre_app=app, pre_season=season, 
                method=method, latent_dimension=latent_dimension, 
                alpha1=alpha1, alpha2=alpha2, alpha3=alpha3, lambda1=lambda1, lambda2=lambda2, lambda3=lambda3, 
                init=init, init_ob=init_ob, uncertainty=uncertainty, k=k, 
                normalization=normalization, random_seed=random_seed)
        for t in range(12):
            AS[fold_num].select(t)
            AS[fold_num].update_ALS_3(t, 5)
    pred = {}
    for t in range(12):
        pred[t] = np.empty((tensor.shape))
        for fold_num in range(10):
            pr = np.einsum('Hr, Ar, Sr -> HAS', AS[fold_num].test_home_matrix[t], 
                           AS[fold_num].app_matrix[t], AS[fold_num].season_matrix[t])
            pred[t][fold_num*28:(fold_num+1)*28] = pr
    mask_tensor = ~np.isnan(tensor)
    
    error = {}
    for t in range(12):
        error[t] = np.sqrt(mean_squared_error(pred[t][:, :, :(t+1)][mask_tensor[:, :, :(t+1)]], tensor[:, :, :(t+1)][mask_tensor[:, :, :(t+1)]]))
    
    
    month_error = {}
    for t in range(12):
        month_error[t] = {}
        for i in range(t, 12):
            month_error[t][i] = np.sqrt(mean_squared_error(pred[i][:, :, t][mask_tensor[:, :, t]], tensor[:, :, t][mask_tensor[:, :, t]]))
    
    app_error = {}
    for a in range(1, 9):
        app_error[a] = {}
        for t in range(12):
            app_error[a][t] = {}
            for i in range(t, 12):
                app_error[a][t][i] = np.sqrt(mean_squared_error(pred[i][:, a, t][mask_tensor[:, a, t]], tensor[:, a, t][mask_tensor[:, a, t]]))
    
    return error, month_error, app_error
```

```python
def get_error(method='active', latent_dimension=3, alpha1=0.1, alpha2=0.1, alpha3=0.1, 
              lambda1=1, lambda2=1, lambda3=1, init='random', init_ob=0, uncertainty='one',
              k=5, normalization='none'):
    error = {}
    month_error = {}
    app_error = {}
    for random_seed in range(5):
        error[random_seed], month_error[random_seed], app_error[random_seed] = active_sensing(method=method, 
                                                                                              latent_dimension=latent_dimension, 
                                                                                              alpha1=alpha1, 
                                                                                              alpha2=alpha2, 
                                                                                              alpha3=alpha3, 
                                                                                              lambda1=lambda1, 
                                                                                              lambda2=lambda2, 
                                                                                              lambda3=lambda3, 
                                                                                              init=init, 
                                                                                              init_ob=init_ob, 
                                                                                              uncertainty=uncertainty,
                                                                                              k=k, 
                                                                                              normalization=normalization, 
                                                                                              random_seed=random_seed)
    m_error = np.empty((5, 12, 12))
    for random_seed in range(5):
        m_error[random_seed] = pd.DataFrame(month_error[random_seed]).values

    a_error = {}
    for a in range(1, 9):
        a_error[a] = np.empty((5, 12, 12))
        for random_seed in range(5):
            a_error[a][random_seed] = pd.DataFrame(app_error[random_seed][a]).values
    return error, m_error, a_error
```

```python
error = {}
m_error = {}
a_error = {}
for method in ['active', 'random']:
    error[method] = {}
    m_error[method] = {}
    a_error[method] = {}
    for normalization in ['none', 'zscore']:
        error[method][normalization] = {}
        m_error[method][normalization] = {}
        a_error[method][normalization] = {}
        for init_ob in [0, 10]:
            print(method, normalization, init_ob)
            error[method][normalization][init_ob],m_error[method][normalization][init_ob], a_error[method][normalization][init_ob] = get_error(method=method, 
                                                                lambda1=lambda1, 
                                                                lambda2=lambda2, 
                                                                lambda3=lambda3, 
                                                                init_ob=init_ob, 
                                                                normalization=normalization)
```
