```python
import numpy as np
import sys
sys.path.append("../code/")
from structure import *
from algo import *
from basic import *
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from autograd.numpy import linalg as LA
import pickle
```

```python
alpha1=0.1
alpha2=0.1
alpha3=0.1
```

```python
init = 'pre'
normalization = 'none'
k = 5
app_matrix = {}
season_matrix = {}
sp = {}
error = {}
for method in ['active']:
    app_matrix[method] = {}
    season_matrix[method] = {}
    sp[method] = {}
    error[method] = {}
    
    for uncertainty in ['one', 'equal', 'weighted']:
        app_matrix[method][uncertainty] = {}
        season_matrix[method][uncertainty] = {}    
        error[method][uncertainty] = {}    

        for random_seed in range(5):
            directory = "../data/fixed_season/"
            app_file = "app-matrix-{}-{}-{}-{}-{}-{}-{}-{}-{}".format(method, init, normalization, uncertainty, alpha1, alpha2, alpha3, k, random_seed)
            season_file = "season-matrix-{}-{}-{}-{}-{}-{}-{}-{}-{}".format(method, init, normalization, uncertainty, alpha1, alpha2, alpha3, k, random_seed)
            error_file = "error-{}-{}-{}-{}-{}-{}-{}-{}-{}".format(method, init, normalization, uncertainty, alpha1, alpha2, alpha3, k, random_seed)
            error_name = directory + error_file
            app_name = directory + app_file
            season_name = directory + season_file

#                     print(e_name)
            try:
                app_file = open(app_name, 'rb')
                season_file = open(season_name, 'rb')
                error_file = open(error_name, 'rb')

                app_matrix[method][uncertainty][random_seed] = pickle.load(app_file)
                season_matrix[method][uncertainty][random_seed] = pickle.load(season_file)
                error[method][uncertainty][random_seed] = pickle.load(error_file)

            except:
                print("No such file!")
                print(app_name)
                print(season_name)
                continue
```
