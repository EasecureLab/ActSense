## tensor factorization

```python
def cost_abs(H, A, S, E_np_masked):
    HAT = np.einsum('Hr, Ar, Sr -> HAS', H, A, S)
    mask = ~np.isnan(E_np_masked)
    error = (HAT - E_np_masked)[mask].flatten()
    return np.sqrt((error ** 2).mean())
```

```python
def cost_abs_2(H, A, S, E_np_masked):
    HAT = np.einsum('Hr, Ar, Sr -> HAS', H, A, S)
    mask = ~np.isnan(E_np_masked)
    error_1 = (HAT - E_np_masked)[mask].flatten() 
    error_2 = 0.01*LA.norm(H) + 0.01*LA.norm(A) + 0.01*LA.norm(S)
    return np.sqrt((error_1 ** 2).mean()) + error_2
```

```python
def factorization(tensor, num_latent, num_iter=2000, lr=0.01, dis=False, random_seed=0, eps=1e-8):
    np.random.seed(random_seed)
    cost = cost_abs_2
    
    args_num = [0, 1, 2]
    mg = multigrad(cost, argnums = args_num)
    M, N, O = tensor.shape

    H = np.random.rand(M, num_latent)
    A = np.random.rand(N, num_latent)
    T = np.random.rand(O, num_latent)

    sum_square_gradients_A = np.zeros_like(A)
    sum_square_gradients_H = np.zeros_like(H)
    sum_square_gradients_T = np.zeros_like(T)
    
    # GD procedure
    for i in range(num_iter):
        del_h, del_a, del_t = mg(H, A, T, tensor)
        
        sum_square_gradients_A += eps + np.square(del_a)
        lr_a = np.divide(lr, np.sqrt(sum_square_gradients_A))
        A -= lr_a * del_a
        
        sum_square_gradients_H += eps + np.square(del_h)
        sum_square_gradients_T += eps + np.square(del_t)

        lr_h = np.divide(lr, np.sqrt(sum_square_gradients_H))
        lr_t = np.divide(lr, np.sqrt(sum_square_gradients_T))

        H -= lr_h * del_h
        T -= lr_t * del_t


        # Projection to non-negative space
        H[H < 0] = 1e-8
        A[A < 0] = 1e-8
        T[T < 0] = 1e-8

        if i % 100 == 0:
            if dis:
                print(cost(H, A, T, tensor))
    
    return H, A, T
```

```python
def get_error(h, a, s, num_latent, tensor):
    mask_tensor = ~np.isnan(tensor)
    h = h.reshape(-1, num_latent)
    a = a.reshape(-1, num_latent)
    s = s.reshape(-1, num_latent)
    hat = np.einsum('Ma, Na, Oa -> MNO', h, a, s)
#     hat = np.einsum('MNa, Oa -> MNO', ha, s)
#     print(hat.shape)
    errors = {}
#     print(hat[:, 3].mean(axis=0))
    
    for i in range(a.shape[0]):
        errors[i] = np.sqrt(mean_squared_error(hat[:, i][mask_tensor[:, i, :s.shape[0]]], tensor[:, i][mask_tensor[:, i, :s.shape[0]]]))
    return errors
```

## ALS

```python
def update_ALS(observed_list, A, b, h, C, d, a, E, f, s):
    num_obs = len(observed_list.keys())
#     print"observations: ", num_obs
    for iters in range(40):
        for i in range(num_obs):
            hid = observed_list[i]['hid']
            aid = observed_list[i]['aid']
            t = observed_list[i]['t']
            e = observed_list[i]['e']

            A[hid] += np.matmul(np.multiply(a[aid], s[t]), np.multiply(a[aid], s[t]).transpose())
            b[hid] += e * (np.multiply(a[aid], s[t]))
            h[hid] = np.matmul(np.linalg.inv(A[hid]), b[hid])
            h[h < 0] = 1e-8
            C[aid] += np.matmul(np.multiply(h[hid], s[t]), np.multiply(h[hid], s[t]).transpose())
            d[aid] += e * (np.multiply(h[hid], s[t]))
            a[aid] = np.matmul(np.linalg.inv(C[aid]), d[aid])
            a[a < 0] = 1e-8
            E[t] += np.matmul(np.multiply(a[aid], h[hid]), np.multiply(a[aid], h[hid]).transpose())
            f[t] += e * (np.multiply(a[aid], h[hid]))
            s[t] = np.matmul(np.linalg.inv(E[t]), f[t])
            s[s < 0] = 1e-8

    
    return A, b, h, C, d, a, E, f, s
```

```python
# create list for the tensor observations
def obs_list(tensor):
    obs = {}
    mask_tensor = ~np.isnan(tensor)
    M, N, O = tensor.shape
    c = 0
    for i in range(M):
        for j in range(N):
            for k in range(O):
                if mask_tensor[i, j, k]:
                    obs[c] = [i, j, k, tensor[i, j, k]]
                    c += 1
    return obs
```

```python
def als_learning_all(tensor, num_latent, random_seed = 0):
    np.random.seed(random_seed)
    mask_tensor = ~np.isnan(tensor)

    # initialization
    num_homes, num_appliance, num_season = tensor.shape
    A = np.empty([num_homes, num_latent, num_latent])
    b = np.empty([num_homes, num_latent, 1])
    h = np.empty([num_homes, num_latent, 1])
    for i in range(num_homes):
        A[i] = np.identity(num_latent)*0.1
        b[i] = np.zeros(num_latent).reshape(-1, 1)
        h[i] = np.random.rand(num_latent).reshape(-1, 1)
        h[i] = np.ones(num_latent).reshape(-1, 1)

    C = np.empty([num_appliance, num_latent, num_latent])
    d = np.empty([num_appliance, num_latent, 1])
    a = np.empty([num_appliance, num_latent, 1])
    for i in range(num_appliance):
        C[i] = np.identity(num_latent)*0.1
        d[i] = np.zeros(num_latent).reshape(-1, 1)
        a[i] = np.random.rand(num_latent).reshape(-1, 1)
        a[i] = np.ones(num_latent).reshape(-1, 1)
        
    E = np.empty([num_season, num_latent, num_latent])
    f = np.empty([num_season, num_latent, 1])
    s = np.empty([num_season, num_latent, 1])
    for t in range(num_season):
        E[t] = np.identity(num_latent)*0.1
        f[t] = np.zeros(num_latent).reshape(-1, 1)
        s[t] = np.random.rand(num_latent).reshape(-1, 1)
        s[t] = np.ones(num_latent).reshape(-1, 1)

    # get the observed list
    observed_list = obs_list(tensor)
    
    for iters in range(5):
        np.random.shuffle(observed_list)
        for i in observed_list:
            hid, aid, sid, e = observed_list[i]

            A[hid] += np.matmul(np.multiply(a[aid], s[sid]), np.multiply(a[aid], s[sid]).transpose())
            b[hid] += e * (np.multiply(a[aid], s[sid]))
            h[hid] = np.matmul(np.linalg.inv(A[hid]), b[hid])
            h[h < 0] = 1e-8
        for i in observed_list:
            hid, aid, sid, e = observed_list[i]
            C[aid] += np.matmul(np.multiply(h[hid], s[sid]), np.multiply(h[hid], s[sid]).transpose())
            d[aid] += e * (np.multiply(h[hid], s[sid]))
            a[aid] = np.matmul(np.linalg.inv(C[aid]), d[aid])
            a[a < 0] = 1e-8
        for i in observed_list:
            hid, aid, sid, e = observed_list[i]
            E[sid] += np.matmul(np.multiply(a[aid], h[hid]), np.multiply(a[aid], h[hid]).transpose())
            f[sid] += e * (np.multiply(a[aid], h[hid]))
            s[sid] = np.matmul(np.linalg.inv(E[sid]), f[sid])
            s[s < 0] = 1e-8
            
    return h, a, s
```

```python
def als_learning_month(tensor, num_latent, random_seed = 0):
    np.random.seed(random_seed)
    mask_tensor = ~np.isnan(tensor)

    num_homes = tensor.shape[0]
    num_appliance = tensor.shape[1]

    A = np.empty([num_homes, num_latent, num_latent])
    b = np.empty([num_homes, num_latent, 1])
    h = np.empty([num_homes, num_latent, 1])
    for i in range(num_homes):
        A[i] = np.identity(num_latent)*0.1
        b[i] = np.zeros(num_latent).reshape(-1, 1)
        h[i] = np.random.rand(num_latent).reshape(-1, 1)
        h[i] = np.ones(num_latent).reshape(-1, 1)

    C = np.empty([num_appliance, num_latent, num_latent])
    d = np.empty([num_appliance, num_latent, 1])
    a = np.empty([num_appliance, num_latent, 1])
    for i in range(num_appliance):
        C[i] = np.identity(num_latent)*0.1
        d[i] = np.zeros(num_latent).reshape(-1, 1)
        a[i] = np.random.rand(num_latent).reshape(-1, 1)
        a[i] = np.ones(num_latent).reshape(-1, 1)
    E = np.empty([12, num_latent, num_latent])
    f = np.empty([12, num_latent, 1])
    s = np.empty([12, num_latent, 1])
    errors = {}
    
    observed_list = {}
    c = 0
    for t in range(12):
        E[t] = np.identity(num_latent)*0.1
        f[t] = np.zeros(num_latent).reshape(-1, 1)
        s[t] = np.random.rand(num_latent).reshape(-1, 1)
        s[t] = np.ones(num_latent).reshape(-1, 1)
        # get the matrix of current month
#         print ("month: ", t+1)
#         print ("observations: ", len(observed_list.keys()))
        for i in range(num_homes):
            for j in range(num_appliance):
                # if reading is not missing
                if mask_tensor[i][j][t]:
                    A[i] += np.matmul(np.multiply(a[j], s[t]), np.multiply(a[j], s[t]).transpose())
                    b[i] += tensor[i][j][t] * (np.multiply(a[j], s[t]))
                    h[i] = np.matmul(np.linalg.inv(A[i]), b[i])
                    h[h < 0] = 1e-8
                    
                    C[j] += np.matmul(np.multiply(h[i], s[t]), np.multiply(h[i], s[t]).transpose())
                    d[j] += tensor[i][j][t] * (np.multiply(h[i], s[t]))
                    a[j] = np.matmul(np.linalg.inv(C[j]), d[j])
                    a[a < 0] = 1e-8
                    
                    E[t] += np.matmul(np.multiply(a[j], h[i]), np.multiply(a[j], h[i]).transpose())
                    f[t] += tensor[i][j][t] * (np.multiply(a[j], h[i]))
                    s[t] = np.matmul(np.linalg.inv(E[t]), f[t])
                    s[s < 0] = 1e-8
                    
                    
                    observed_list[c] = {}
                    observed_list[c]['hid'] = i
                    observed_list[c]['aid'] = j
                    observed_list[c]['t'] = t
                    observed_list[c]['e'] = tensor[i,j,t]
                    c += 1          
                    
        A, b, h, C, d, a, E, f, s = update_ALS(observed_list, A, b, h, C, d, a, E, f, s)
    return h, a, s
```
