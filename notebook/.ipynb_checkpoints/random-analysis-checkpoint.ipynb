```python
import sys
sys.path.append("../code")
from basic import *
from algo import *
import numpy as np
import pickle 
import pandas as pd
%matplotlib inline
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2
```

```python
# for fixed season factor, latent=3, k=5
random_pred = pickle.load(open("../data/fixed_season/random/2015/pred-pre-5-3", 'rb'), encoding='latin1')
active_pred = pickle.load(open("../data/fixed_season/active/2015/pred-active-pre-none-one-0.1-0.1-0.1-5-3-49", 'rb'), encoding='latin1')
equal_pred = pickle.load(open("../data/fixed_season/active/2015/pred-active-pre-none-equal-0.1-0.1-0.1-5-3-49", 'rb'), encoding='latin1')
```

```python
mean = {}
mean['active'] = active_pred.mean(axis=0)
mean['random'] = random_pred.mean(axis=0)
mean['active equal'] = equal_pred.mean(axis=0)
```

```python
tensor, hid = get_tensor(2015, 'filter')
```

```python
# get rmse
mask_tensor = ~np.isnan(tensor)
accu_rmse = {}
month_rmse = {}
app_rmse = {}
for app in range(1, 9):
    app_rmse[app] = {}
for method in ['random', 'active', 'active equal']:
    accu_rmse[method] = {}
    month_rmse[method] = {}
    app_rmse[app][method] = {}
    for t in range(12):
        accu_rmse[method][t] = np.sqrt(mean_squared_error(mean[method][t, :, 1:, :(t+1)][mask_tensor[:, 1:, :(t+1)]], 
                                                  tensor[:, 1:, :(t+1)][mask_tensor[:, 1:, :(t+1)]]))
        month_rmse[method][t] = np.sqrt(mean_squared_error(mean[method][t, :, 1:, t][mask_tensor[:, 1:, t]], 
                                                  tensor[:, 1:, t][mask_tensor[:, 1:, t]]))
        for app in range(1, 9):
            app_rmse[method][app][t] = np.sqrt(mean_squared_error(mean[method][t, :, app, t][mask_tensor[:, app, t]], 
                                                  tensor[:, app, t][mask_tensor[:, app, t]]))
```
