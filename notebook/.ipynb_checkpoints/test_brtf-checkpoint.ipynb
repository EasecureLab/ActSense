```python
def BTCP_TC(Y, mask_tensor, iters, display=False, random_seed=0):

    np.random.seed(random_seed)

    # mask_tensor = ~np.isnan(Y) * 1
    nObs = mask_tensor.sum()
    M, N, P = Y.shape
    Y = Y * mask_tensor
    print(M, N, P)
    R = 3
    initVar = 1

    # initialization
    a_gamma0 = 1e-6
    b_gamma0 = 1e-6
    a_beta0 = 1e-6
    b_beta0 = 1e-6
    a_alpha0 = 1e-6
    b_alpha0 = 1e-6

    gammas = (a_gamma0 + np.spacing(1)) / (b_gamma0 + np.spacing(1)) * np.ones(R)
    beta = (a_beta0 + np.spacing(1)) / (b_beta0 + np.spacing(1))
    alphas = np.ones(Y.shape) * ((a_alpha0 + np.spacing(1)) / (b_alpha0 + np.spacing(1))) / initVar

    E = alphas * 0.5 * np.random.rand(M, N, P)
    Sigma_E = (1 / alphas) * np.ones(Y.shape)

    # random initialization
    LB = {}
    Z = {}
    ZSigma = {}
    dvar = np.sqrt((Y[mask_tensor] ** 2).mean())

    # Z[0], Z[1], Z[2] = factorization(Y, num_latent=3)
    for i in range(3):
        Z[i] = np.random.rand(Y.shape[i], R) * dvar
        ZSigma[i] = np.repeat(np.diag(1 / gammas)[:, :, np.newaxis], Y.shape[i], axis=2)

    # Model Learning
    EZZT = {}
    for i in range(3):
        EZZT[i] = np.reshape(ZSigma[i], (R * R, -1)).T + tensorly.tenalg.khatri_rao([Z[i].T, Z[i].T]).T
    print(EZZT)
    for it in range(1):
        
        Aw = np.diag(gammas)

        t_mask = tensor.tensor(mask_tensor)
        ye_tensor = tensor.tensor((Y - E) * mask_tensor)

        # update parameters for U matrix
        ENZZT = np.reshape(tensorly.tenalg.khatri_rao([EZZT[2], EZZT[1]]).T @ tenmat.tenmat(t_mask, rdim=[0], cdim=[2, 1]).tondarray().T, (R, R, -1))
        FslashY = tensorly.tenalg.khatri_rao([Z[2], Z[1]]).T @ tenmat.tenmat(ye_tensor, rdim=[0], cdim=[2, 1]).tondarray().T
#         print(FslashY.shape)
        for i in range(M):
            ZSigma[0][:, :, i] = np.linalg.inv(beta * ENZZT[:, :, i] + Aw)
            Z[0][i, :] = (beta * ZSigma[0][:, :, i] @ FslashY[:, i]).T
        EZZT[0] = (np.reshape(ZSigma[0], (R * R, -1)) + tensorly.tenalg.khatri_rao([Z[0].T, Z[0].T])).T
#         print("iter {}".format(it))
#         print(FslashY[:, 5])
        # update parameters for V matrix
        ENZZT = np.reshape(tensorly.tenalg.khatri_rao([EZZT[2], EZZT[0]]).T @ tenmat.tenmat(t_mask, rdim=[1], cdim=[2, 0]).tondarray().T, (R, R, -1))
        FslashY = tensorly.tenalg.khatri_rao([Z[2], Z[0]]).T @ tenmat.tenmat(ye_tensor, rdim=[1], cdim=[2, 0]).tondarray().T
        for i in range(N):
            ZSigma[1][:, :, i] = np.linalg.inv(beta * ENZZT[:, :, i] + Aw)
            Z[1][i, :] = (beta * ZSigma[1][:, :, i] @ FslashY[:, i]).T
        EZZT[1] = (np.reshape(ZSigma[1], (R * R, -1)) + tensorly.tenalg.khatri_rao([Z[1].T, Z[1].T])).T
#         print(FslashY)
        # update parameters for T matrix
        ENZZT = np.reshape(tensorly.tenalg.khatri_rao([EZZT[1], EZZT[0]]).T @ tenmat.tenmat(t_mask, rdim=[2], cdim=[1, 0]).tondarray().T, (R, R, -1))
        FslashY = tensorly.tenalg.khatri_rao([Z[1], Z[0]]).T @ tenmat.tenmat(ye_tensor, rdim=[2], cdim=[1, 0]).tondarray().T
        for i in range(P):
            ZSigma[2][:, :, i] = np.linalg.inv(beta * ENZZT[:, :, i] + Aw)
            Z[2][i, :] = (beta * ZSigma[2][:, :, i] @ FslashY[:, i]).T
        EZZT[2] = (np.reshape(ZSigma[2], (R * R, -1)) + tensorly.tenalg.khatri_rao([Z[2].T, Z[2].T])).T

        # update latent tensor X
        X = np.einsum('Hr, Sr, Kr -> HSK', Z[0], Z[1], Z[2])

        # update hyperparameters gamma
        a_gammaN = (0.5 * (M + N + P) + a_gamma0) * np.ones(R)
        b_gammaN = 0
        for i in range(3):
            b_gammaN += np.diag(Z[i].T @ Z[i]) + np.diag(ZSigma[i].sum(axis=2))
        b_gammaN = b_gamma0 + 0.5 * b_gammaN
        gammas = a_gammaN / b_gammaN

        # update noise beta
        temp1 = {}
        EX2 = 0
        for i in range(R):
            for n in range(3):
                temp1[n] = EZZT[n][:, i * 3:(i + 1) * 3]
#             print((mask_tensor.reshape(1, -1) @ tensorly.tenalg.khatri_rao([temp1[2], temp1[1], temp1[0]])))
            EX2 += (mask_tensor.reshape(1, -1) @ tensorly.tenalg.khatri_rao([temp1[2], temp1[1], temp1[0]])).sum()
#         print(EX2)
        EE2 = ((E ** 2 + Sigma_E) * mask_tensor).sum()
        err = Y.flatten().T @ Y.flatten() - 2 * Y.flatten().T @ X.flatten() - 2 * Y.flatten().T @ E.flatten() + 2 * X.flatten().T @ E.flatten() + EX2 + EE2
        a_betaN = a_beta0 + 0.5 * nObs
        b_betaN = b_beta0 + 0.5 * err
        beta = a_betaN / b_betaN

        # Update Sparse matrix E
        Sigma_E = np.divide(mask_tensor * 1.0, (alphas + beta))
        E = (beta * Sigma_E * (Y - X) * mask_tensor)

        # update alphas
        inf_flag = 2
        if inf_flag == 1:
            a_alphaN = a_alpha0 + 0.5 * (mask_tensor * 1.0)
            b_alphaN = b_alpha0 + 0.5 * (E ** 2 + Sigma_E)
            alphas = a_alphaN / b_alphaN
        else:
            a_alphaN = a_alpha0 + 1 - alphas * Sigma_E
            b_alphaN = b_alpha0 + E ** 2 + np.spacing(1)
            alphas = a_alphaN / b_alphaN

        if display:
            print(it, np.sqrt((Y - X) ** 2).mean())
#         print(b_betaN, a_betaN, b_gamma0, beta, err, EX2)

    for i in range(1, 7):
        if Z[1][i, :].sum() == 0:
            print("{} is not queried", i)
            Z[1][i, :] = np.random.rand(R)

    return Z, ZSigma
```
