```python
def get_train_test(tensor, homeids, num_folds=5, fold_num=0):
    homeids = np.array(homeids)
    num_homes= tensor.shape[0]
    k = KFold(n_splits=num_folds)
    train, test = list(k.split(range(0, num_homes)))[fold_num]
    tr_ids, tt_ids = list(k.split(range(0, num_homes)))[fold_num]
    print(train)
    return tensor[train], tensor[test], list(homeids[tr_ids]), list(homeids[tt_ids])
```

```python
def get_uncertainty(s, AS_equal):
    ut = {}
    sum_ut = {}
#     season_id = 0
    ut = np.zeros((AS_equal.num_appliance, AS_equal.num_train, 11, 3))
    sum_ut = np.zeros((AS_equal.num_appliance, AS_equal.num_train, 11))
    sa = np.zeros((AS_equal.num_appliance, 11, 3, 1))
    sh = np.zeros((AS_equal.num_train, 11, 3, 1))
    for t in range(s, 11):
        for i in range(AS_equal.num_train):
            for j in range(AS_equal.num_appliance):
                h_circ_a = np.multiply(AS_equal.homes[i].h, AS_equal.apps[j].a)
                h_circ_s = np.multiply(AS_equal.homes[i].h, AS_equal.seasons[t].s)
                a_circ_s = np.multiply(AS_equal.apps[j].a, AS_equal.seasons[t].s)
                sa[j][t] = a_circ_s
                sh[i][t] = h_circ_s

                var_1 = np.sqrt(np.matmul(np.matmul(a_circ_s.transpose(), AS_equal.homes[i].AInv), a_circ_s))
                var_2 = np.sqrt(np.matmul(np.matmul(h_circ_s.transpose(), AS_equal.apps[j].CInv), h_circ_s))
                var_3 = np.sqrt(np.matmul(np.matmul(h_circ_a.transpose(), AS_equal.seasons[season_id].EInv), h_circ_a))

                ut[j][i][t][0] = var_1[0][0]
                ut[j][i][t][1] = var_2[0][0]
                ut[j][i][t][2] = var_3[0][0]
                sum_ut[j][i][t] = var_1[0][0] + var_2[0][0] + var_3[0][0]
    return ut, sum_ut
```
