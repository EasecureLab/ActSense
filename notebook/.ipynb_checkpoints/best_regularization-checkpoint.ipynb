```python
init = 'pre'
uncertainty = 'one'
normalization = 'none'
result = {}

for method in ['active', 'random', 'all']:
    result[method] = {}
    for init, iters in [('pre', 1), ('random', 5)]:
        result[method][init] = {}
        for lambda_ in [10000]:
            lambda_ = float(lambda_)
            result[method][init][lambda_] = {}
            
            for random_seed in range(5):
                directory = "../data/errors/"
                e_file = "error-{}-{}-{}-{}-{}-{}".format(method, lambda_, init, iters, normalization, random_seed)
                m_file = "m-err-{}-{}-{}-{}-{}-{}".format(method, lambda_, init, iters, normalization, random_seed)
                a_file = "a-err-{}-{}-{}-{}-{}-{}".format(method, lambda_, init, iters, normalization, random_seed)
                am_file = "am-err-{}-{}-{}-{}-{}-{}".format(method, lambda_, init, iters, normalization, random_seed)
                sm_file = "sm-err-{}-{}-{}-{}-{}-{}".format(method, lambda_, init, iters, normalization, random_seed)
                sp_file = "sp-err-{}-{}-{}-{}-{}-{}".format(method, lambda_, init, iters, normalization, random_seed)
                name = directory + e_file
#                     print(name)
                try:
                    file = open(name, 'rb')
                    result[method][init][lambda_][random_seed] = pickle.load(file) 
                except:
                    CMD = 'python active-sensing.py {} none {} {} {} {} &'.format(method, init, iters, lambda_, random_seed)
                    print(CMD)
```

```python
fig, axes = plt.subplots(ncols=2, nrows = 30, figsize=(12, 90))
ax = axes.flatten()
c = 0

for method in ['active', 'random', 'all']:
    for init in ['pre']:
        for lambda_ in [10, 100, 1000, 10000]:
            lambda_ = float(lambda_)
            df = None
            for iters in [1, 5, 10, 20]:
                df_t = pd.DataFrame(result[method][init][lambda_][iters]).mean(axis=1)
                df = pd.concat([df, df_t], axis=1)
            df.plot(kind='bar', ax=ax[c], title = "method = {}, init = {}, lambda_ = {}".format(method, init, lambda_))
            ax[c].legend([1, 5, 10, 20], loc='upper right')
            c += 1
plt.subplots_adjust(hspace=0.5)
```
