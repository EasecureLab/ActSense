### Set up parameters

```python
year = 2015
dataset = 'artificial'
init = 'pre'
k = 5
latent_dimension = 4
reg=True
if dataset == 'artificial':
    order = APPLIANCE_ORDER_OBSERVED
else:
    order = APPLIANCE_ORDER_MISSING
```

### get the results of active selection

```python
result = {}
result['random'] = get_active_results(year=year, method='random', k=k, latent_dimension=latent_dimension, reg=reg)
result['qbc'] = get_active_results(year=year, method='qbc', k=k, latent_dimension=latent_dimension, reg=reg)
result['prev_future_weighted'] = get_active_results(year=year, uncertainty='prev_future_weighted', k=k, latent_dimension=latent_dimension, reg=reg)
```

## get the accumulated Error of random method and active method

```python
error = {}
for e_type in ['overall', 'app']:
    error[e_type] = {}
    for t_type in ['accu', 'month']:
        error[e_type][t_type] = {}
        for param in result.keys():
            error[e_type][t_type][param] = {}
            for idx, appliance in enumerate(order[1:]):
                error[e_type][t_type][param][appliance] = {}
                for t in range(12):
                    error[e_type][t_type][param][appliance][t] = {}
                    for random_seed in range(10):
                        error[e_type][t_type][param][appliance][t][random_seed] = {}
```

### using RMSE as the evaluation metric

```python
tensor, homeidx = get_tensor(year, dataset)
mask_tensor = ~np.isnan(tensor)
```

```python
for param in result.keys():
    for t in range(12):
        # random method
        for idx, appliance in enumerate(order[1:]):
            idx += 1
            for random_seed in range(10):
                # for active method
                predApp = result[param]['prediction'][random_seed][t][:, idx, :] # tensor
                predApp[predApp<0] = 0
                
                # accumulated appliance error
                appAccuPred = predApp[:, :(t+1)]
#                 print(appliance, appAccuPred.min())
                error['app']['accu'][param][appliance][t][random_seed] = np.sqrt(mean_squared_error(appAccuPred[mask_tensor[:, idx, :(t+1)]], 
                                                                                               tensor[:, idx, :(t+1)][mask_tensor[:, idx, :(t+1)]]))
                # appliance error in current month
                appMonthPred = predApp[:, t]
                error['app']['month'][param][appliance][t][random_seed] = np.sqrt(mean_squared_error(appMonthPred[mask_tensor[:, idx, t]], 
                                                                                                tensor[:, idx, t][mask_tensor[:, idx, t]]))
                
                
```

```python
# pec error: energy correctly assigned
new_error = {}
for param in result.keys():
    new_error[param] = {}
    for t in range(12):
        new_error[param][t] = {}
        for idx, appliance in enumerate(order[1:]):
            new_error[param][t][appliance] = {}
            idx += 1
            for random_seed in range(10):
                
                predApp = result[param]['prediction'][random_seed][t][:, idx, :]
                predApp[predApp < 0] = 0
                appAccuPred = predApp[:, :(t+1)]
                
                pred_frac = (appAccuPred[mask_tensor[:, idx, :(t+1)]] / tensor[:, 0, :(t+1)][mask_tensor[:, 0, :(t+1)]]) * 100
                gt_frac = (tensor[:, idx, :(t+1)][mask_tensor[:, idx, :(t+1)]] / tensor[:, 0, :(t+1)][mask_tensor[:, 0, :(t+1)]]) * 100
                
#                 pec = np.abs(pred_frac - gt_frac) * 100
                
                rmspec = np.sqrt(mean_squared_error(pred_frac, gt_frac))
                new_error[param][t][appliance][random_seed] = rmspec
```

```python
def calculate_selected_appliance(sp, k):
    selected_app = np.zeros((12, len(order)))
    for random_seed in range(10):
        for fold_num in range(5):
            for t in range(1, 12):
                for i in range(k):
#                     print(random_seed, fold_num, t, i)
                    app_idx = sp[random_seed][fold_num][t][i][1]
                    selected_app[t][app_idx] += 1
    return selected_app
```
